```python
from hmmlearn.hmm import GaussianHMM
import pandas as pd
import numpy as np
import math
from collections import Counter
```

```python
def shrink_array(array,size):
    
    ratio = float(len(array)) / float(size)
    res = []
    for i in range(size):
        res.append(np.mean(array[math.floor(i*ratio):math.ceil((i+1.0)*ratio)]))
    return res
```

```python
sensor_data = pd.DataFrame.from_csv("../../data/database/sensor_data.csv")
```

```python
sensor_data_acc = sensor_data[sensor_data.SENSORTYPE == 1]
sensor_data_acc_tag12 = sensor_data_acc[(sensor_data_acc.TagName == 'Tag1') | (sensor_data_acc.TagName == 'Tag2')]
```

```python
tag_id_groupby = sensor_data_acc_tag12.groupby(['TagName', 'tester_id'])
```

```python
tag_id_dict = tag_id_groupby.groups
```

```python
y = list(tag_id_dict.keys())
y = [i[0] for i in y]
```

```python
acc = np.zeros(shape=(0,200,2))
# acc_x = np.zeros(shape=(0,200))
# acc_y = np.zeros(shape=(0,200))


for k in list(tag_id_dict.keys()):
    cur_group = tag_id_groupby.get_group(k)
    cur_x = shrink_array(cur_group['VALUES1'].values, 200) # a 1d array with length 200
    cur_y = shrink_array(cur_group['VALUES2'].values, 200)
    cur_xy = [list(i) for i in list(zip(cur_x,cur_y))] # shape (200,2)
    acc = np.append(acc, np.array([cur_xy]), axis=0)
```
